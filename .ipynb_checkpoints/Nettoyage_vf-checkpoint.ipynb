# Chargement des librairies et du dataset

Importation des librairies nécessaires :

```python
import pandas as pd
import matplotlib as matplot
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set(style="whitegrid")
from collections import Counter
```

Chargement du dataset :

```python
data = pd.read_csv('data_openfoodfacts.csv', sep='\t')
```

```python
print(data.info())
```

Le dataset contient 888 151 entrées, décrites par 175 variables. 
Parmi ces variables, une majorité sont numériques (principalement les données nutritionelles pour 100g et le reste est de type object et sont des chaines de caractères (URL, description ...).

# Nettoyage

## Opérations de base

Pour la suite, afin d'alléger le dataset, on retire certaines variables qui apportent peu sur la connaissance de la nature du produit et de sa composition (infos database, variables redondantes identifiées via le suffixe 'en' ou 'tags' dont on conserve un seul exemplaire, packaging). La liste des variables retirées est disponible dans le fichier 'liste_variables_a_retirer'.

```python
variables_to_drop = pd.read_csv('liste_variables_a_retirer', header=None)[0].tolist()
data.drop(labels=variables_to_drop, axis=1, inplace=True)
```

```python
print(data.shape)
print(data.columns[:17])
```

Au total ce sont 50 variables qui ont été retirées, afin de conserver uniquement les données descriptives essentielles pour connaitre la nature du produit (visibles ci -essus : nom, marque, nature ...) et ses données nutrionnelles.

On retire ensuite les entrées dupliquées et les variables vides s'il y en a :

```python
data.dropna(axis=1, how='all', inplace=True)
print(data.shape)
```

9 variables ont été retirées suite à cette manipulation.

```python
data.drop_duplicates(keep = 'first', inplace = True) 
print(data.shape)
```

36 entrées ont été retirées suite à cette manipulation, la dataset contient peu de duplicats.

## Zoom sur la France

L'objectif du projet étant de proposer une idée d'application aux services de santé publique français, on conservera uniquement les produits disponibles en France, si ceux-ci sont suffisamment représentés.

```python
fr_filter = data['countries_tags'].astype(dtype='str').str.split(',')
countries_count = fr_filter.apply(lambda lo: Counter(lo)).sum()
countries_count_ordered = sorted(countries_count.items(), key=lambda kv: kv[1], reverse=True)
countries_count_ordered = pd.DataFrame(countries_count_ordered, columns=['country', 'entries'])
countries_count_ordered.set_index(keys='country', inplace=True)
countries_count_ordered = countries_count_ordered / data.shape[0] * 100
countries_count_ordered['cumulated_percentage'] = countries_count_ordered['entries'].cumsum()
print(countries_count_ordered.head())
```

```python
fig, ax = plt.subplots(figsize=(8,5))
countries_count_ordered[:7].plot.bar(y='entries', ax=ax, legend=None)
plt.plot(countries_count_ordered['cumulated_percentage'],c="red",marker='o')
ax.set_ylim(0,100)
ax.set_title('Product availability by country', fontsize=25)
ax.set_ylabel('Product percentage (%)', fontsize=20)
ax.set_xlabel('')
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
#plt.savefig( 'Product_country_availability.png', format='png', bbox_inches = "tight", dpi=600)
```

Plus de 60% des produits sont disponibles en France, il s'agit du pays le plus représenté, suivit par les US avec 20% et d'autres pays européens à moins de 5% chacun. Conserver unqiuement les produits français fait donc sens.

```python
fr_filtered = data[fr_filter.apply(lambda lo: 'en:france' in lo)]
print(fr_filtered.shape)
```

315 149‬ produits ont été retirés suite à cette opération.

Suite à l'analyse des 4 variables précédentes, on voit que la variable 'pnns_groups_2' est la plus intéressante d'un point de vue description du produit car la plus détaillée et la plus homogène (car standardisée). Néanmoins ce n'est pas celle qui permet de conserver le plus d'individus.
En effet on voit que l'on aura au mieux 206 285 entrées disponibles sur les 572 966 du dataset filtré sur les produits français avec cette variable.
La variable 'main_category_en' elle permet d'augmenter ce nombre à 233 888, permettant un gain de 13% de données disponibles.

```python
categories = fr_filtered.loc[:,['main_category_en', 'pnns_groups_1', 'pnns_groups_2']].replace('unknown', np.nan)
print(categories.info())
```

```python
print(categories[categories.notna().any(axis=1)].shape)
```

### Consevation des données dont on connait le type uniquement : 

```python
data_filtered = fr_filtered.loc[fr_filtered['pnns_groups_2'] != 'unknown'].dropna(axis=0, subset=['pnns_groups_2'])
print(data_filtered.shape)
data_filtered['pnns_groups_2'] = data_filtered['pnns_groups_2'].str.lower().astype('category')
print(data_filtered.shape)
data_filtered.dropna(axis=1, how='all', inplace=True)
print(data_filtered.shape)
data_filtered.drop(labels=['nova_group','pnns_groups_1','main_category_en'], axis=1, inplace=True)
print(data_filtered.shape)
```

Suite à cette opération on se retrouve avec un peu plus de 200 000 produits, ce qui semble suffisant pour effectuer de première analyse. Dans le cas où l'on aurait souhaité par la suite réaliser un modèle, il aurait par contre été intéressant de combiner 'pnns_groups_2' et 'main_category_en' pour maximiser le nombre de données disponibles. On en profite également pour retirer les variables 'nova_group', 'pnns_groups_1' et 'main_category_en' qui ne sont plus utiles.

## Gestion des variables peu représentées :

Pour continuer à nettoyer le dataset, on va s'intéresser aux variables peu représentées.
Pour cela le dataset sera séparé en deux parties : les variables descriptives et les variables nutritionnelles pour 100g.
On se concentrera sur les données pour 100g pour calculer ensuite le pourcentage de données non assignées.

```python
infos = data_filtered.loc[:,:'pnns_groups_2']
nutrition =data_filtered.loc[:,'energy_100g':]
print(infos.shape)
print(nutrition.shape)
```

```python
null_entries = (nutrition.shape[0] - nutrition.count()) / nutrition.shape[0] * 100
```

```python
fig3, ax3 = plt.subplots(figsize=(5,40))
sns.barplot(null_entries.values, null_entries.index, label="% de données non assignées", color="b", ax=ax3)
ax3.set_title('% de données non assignées', fontsize=25)
ax3.xaxis.set_major_locator(matplot.ticker.MultipleLocator(10))
ax3.set_xlim(0,100)
plt.yticks(fontsize=15)
#plt.savefig('%_donnees_manquantes.png', format='png', bbox_inches = "tight", dpi=600)
```

On constate que de nombreuses variables ont un nombre très élevé de valeurs inconnues.
Nous allons nous concentrer sur les variables ayant un taux de données non renseignées inférieur à 80%.

```python
fig4, ax4 = plt.subplots(figsize=(5,5))
sns.barplot(null_entries[null_entries < 80].values, null_entries[null_entries < 80].index,
            label="% de données non assignées", color="b", ax=ax4)
ax4.set_title('% de données non assignées', fontsize=20)
ax4.xaxis.set_major_locator(matplot.ticker.MultipleLocator(10))
ax4.set_xlim(0,100)
plt.yticks(fontsize=15)
#plt.savefig('variables_moins_de_80%_de_NaN.png', format='png', bbox_inches = "tight", dpi=600)
```

On retrouve les informations nutritionnelles qui sont conventionnellement renseignées sur les emballages.
Les fibres ont tout de même un haut taux de valeurs non assignées (environ 65%) car celles-ci ne font pas partie des éléments qu'il est impératif de renseigner.
Afin de traiter le dataset, nous allons nous concentrer sur ces données en particulier, les autres étant jugées comme très spécifiques. Néanmoins afin de ne pas perdre trop d'informations (les variables très peu représentées peuvent être une spécificité du produit et un élément distinctif), nous allons conserver la présence de ces éléments dans un nouvelle variable nommée 'additionnals_infos_100g'.

```python
specific_components = nutrition.loc[:,null_entries[null_entries > 80].index].isnull()
additional_infos_100g = specific_components.apply(lambda x: str(x[x==False].index.values), axis=1)
print(additional_infos_100g.head())
```

On ajoute cette nouvelle variable au dataframe 'infos' qui contient les données plus générales :

```python
infos['additional_infos_100g'] = additional_infos_100g
print(infos.shape)
```

et l'on retire ensuite les variables avec un pourcentage de valeurs non assignées inférieures à 80% du dataset 'nutrition'.

```python
nutrition.drop( labels=null_entries[null_entries > 80].index, inplace=True, axis=1)
print(nutrition.shape)
```

Au final, on aura ajouté une variable pour le dataframe 'infos' et retiré 87 du dataframe 'nutrition'.
On combine ces deux dataframe pour obtenir notre dataframe de travail contenant 206 285 produits et 29 variables.

```python
data_simplified = pd.concat([infos, nutrition], axis=1)
print(data_simplified.shape)
```

## Gestion des produits ayant des données non assignées :

Dans la partie précédente les variables intéressantes suivantes ont été identifées :

'energy_100g','fat_100g','saturated_fat_100g','carbohydrates_100g','sugars_100g','proteins_100g','salt_100g','sodium_100g','nutrition_score_fr_100g', 'nutrition_score_uk_100g' , 'fiber_100g'

Hormis pour les fibres, le reste constitue des données nutritionnelles essentielles qui sont obligatoires lors de l'étiquetage d'un produit, il n'y a donc pas de raison qu'une de ces valeurs ne soient pas présente, si ce n'est une erreur de saisie.
On retira donc les produits pour lesquels il manque une de ces valeurs.

Le cas des fibres étant un peu particulier, car sa mention elle n'est pas obligatoire, sera traité différemment par la suite.

Au passage, certaines variables contiennent des tirets, on les remplacera par des underscore pour éviter d'éventuels problèmes d'interprétation avec le signe '-' lors d'opérations.

```python
data_simplified.columns = data_simplified.columns.str.replace('-','_')
print(data_simplified.columns)
```

### Cas des données nutritionelles essentielles

```python
data_simplified.dropna( subset=['energy_100g','fat_100g','saturated_fat_100g','carbohydrates_100g','sugars_100g',
                                'proteins_100g','salt_100g','sodium_100g','nutrition_score_fr_100g', 'nutrition_score_uk_100g'], inplace=True)
print(data_simplified.shape)
```

En retirant les produits ayant une données manquantes sur une des données nutritionelles, on passe de 206 285 produits à 170 718. 

### Cas des fibres

En ce qui concerne les fibres, les données manquantes sont plus nombreuses (car non soumises à obligation de déclaration). Néanmoins il est également possible qu'un aliment ne contienne pas de fibres, on va donc regarder les catégories d'aliments pour lesquels cette donnée n'est pas renseignée.

```python
fibre = data_simplified[data_simplified['fiber_100g'].isnull()]
display(fibre['pnns_groups_2'].value_counts())
```

On voit que les types d'aliments les plus représentés sont d'origine animale (viande, fromage, poisson, lait ...) ce qui est normal. Néanmoins on observe que la donnée 'fiber_100g' est manquante dans une proportion importante pour des céréales et des légumes qui sont eux normalement riches en fibres et pour lesquels la données devrait être présente.
Afin de palier à ce problème, on va compléter les données manquantes de fibres pour un produit par la médiane (moins sujette aux outliers) de la teneur en fibre pour 100g de sa catégorie d'aliment.

```python
data_simplified['fiber_100g'] = data_simplified.groupby(
    'pnns_groups_2')['fiber_100g'].transform(lambda group: group.fillna(group.median()))
```

```python
print(data_simplified.info())
```

On obtient donc 170 718 produits exploitables sur leurs données nutritionnelles essentielles.

## Gestion des données aberrantes :

Maintenant que l'on a obtenu un dataset avec des données nutritionnelles complètes, on va regarder une nouvelle fois les statistiques descriptives pour chacune d'entre elles.

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(data_simplified.describe())
```

On constate plusieurs anomalies : par exemple une énergie pour 100g apportant 22 740 kJ, soit presque trois fois les apports journaliers nécessaires pour une homme adulte (8400kJ) ou bien encore une teneur en sucre pour 100g supérieure à 100g. Afin de corriger cela on va supprimer les produits dont l'énergie pour 100g dépasse les AJR d'un adulte ou si l'une des données nutritionnelles pour 100g dépasse les 100g.

```python
data_cleaned= data_simplified[(data_simplified['energy_100g']<8400) & 
              (data_simplified.loc[:,['fat_100g','saturated_fat_100g', 'carbohydrates_100g', 'sugars_100g', 
                                    'fiber_100g', 'proteins_100g', 'salt_100g', 'sodium_100g']] < 100).all(axis=1)]
print(data_cleaned.shape)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(data_cleaned.describe())
```

On constate que certaines valeurs très extrêmes restent présentes (92g de protéines pour 100g de produit par exemple). Nous allons regarder quels sont ces produits pour voir s'il s'agit possiblement d'une erreur.

```python
print(data_cleaned.loc[data_cleaned['salt_100g'].idxmax(),['product_name','pnns_groups_2']])
```

```python
print(data_cleaned.loc[data_cleaned['fat_100g'].idxmax(),['product_name','pnns_groups_2']])
```

```python
print(data_cleaned.loc[data_cleaned['fiber_100g'].idxmax(),['product_name','pnns_groups_2']])
```

```python
print(data_cleaned.loc[data_cleaned['proteins_100g'].idxmax(),['product_name','pnns_groups_2']])
```

En regardant ces 4 variables on distingue deux cas :

- Des erreurs, comme pour 'fiber_100g' et 'salt_100g' car un steack haché ne contient pas de fibre (le consommateur a sans doute confondu fibre alimentaire et fibre musculaire) et la limonade ne contient pas autant de sel
- Des produits atypiques quasi mono composant comme l'huile d'olive qui est pure matière grasse ou bien encore le complément alimentaire qui est en grande partie de la protéine.

Pour la suite, on essaiera de réduire la nombre de ces erreurs et de ces produits atypiques à l'aide du Z score.

## Test de réduction d'outliers à l'aide du zscore

Dans cette partie on essaiera de limiter la présence de produits très atypiques ou bien encore d'erreurs en calculant le Zscore pour chaque composante nutrionnelle d'un produit, en le positionnant uniquement par rapport à sa catégorie. De manière conventionelle, on conservera les produits pour lesquels aucune des données nutritionelles n'affiche un zscore supérieur à 3.

```python
def z_score(df): return (df-df.mean())/df.std(ddof=0)
```

```python
data_zscored = data_simplified.loc[:,['pnns_groups_2','fat_100g','saturated_fat_100g', 'carbohydrates_100g', 'sugars_100g', 
                                    'fiber_100g', 'proteins_100g', 'salt_100g', 'sodium_100g']].groupby(
    'pnns_groups_2').transform(lambda group: z_score(group))
```

```python
display(data_zscored.head())
```

vérification du calcul du score en le calculant sur un DF avec une seule carégorie :

```python
test = data_simplified.loc[:,['pnns_groups_2','fat_100g','saturated_fat_100g', 'carbohydrates_100g', 'sugars_100g','fiber_100g', 'proteins_100g', 'salt_100g', 'sodium_100g']]
test = test.loc[test['pnns_groups_2'] == 'fruits']
test.drop(labels='pnns_groups_2', axis=1, inplace=True)
display(z_score(test).head())
```

En comparant avec l'entrée n°25 du dataframe z_scored, on voit que le calcul s'est bien effectué par catégorie.

```python
test_cleaned = data_cleaned[(data_zscored < 3).all(axis=1)]
print(test_cleaned.shape)
display(test_cleaned.describe())
```

Après application, on voit que certains produits atypiques comme le complémentaire alimentaire en protéines ou bien les erreurs sur la fibre et le sel mentionnées précédemment ont été suprimmés, ce qui devrait permettre de mieux comparer les produits entre eux par la suite.
