## Recherche de corrélations linéaires via le coefficient corrélation de Pearson

```python
pearson_correlation = nutrition.corr()
mask = np.zeros_like(pearson_correlation, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
cmap = sns.color_palette("coolwarm", 20)
fig1, ax1 = plt.subplots(figsize=(15,15))
ax1 = sns.heatmap(pearson_correlation, mask=mask, cmap=cmap, vmin=-1, vmax=1, 
            linewidths=1, square=True, annot=True,annot_kws={"size": 16})
ax1.tick_params(labelsize=18)
ax1.collections[0].colorbar.ax.tick_params(labelsize=20)
ax1.set_title('Pearson correlation coefficient', fontsize=25)
#plt.savefig( 'pearson_correlation__variables_importantes.png', format='png', bbox_inches = "tight", dpi=400)
```

On observe les points suivants :
- Le sodium et le sel sont complètement corrélés positivement, ce qui parait normal étant donné que le sel (Chlorure de sodium) est en partie composé de sodium.
- Les sucres et les glucides sont aussi assez fortement corrélés positivement, les sucres étant une sous catégorie des glucides. On peut faire le même constat pour les graisses saturées et la matière grasse.
- Les fibres sont modérément corrélées aux glucides. Cela s'explique par le fait que l'on trouve des fibres dans les féculents qui sont aussi très riches en glucides mais elles sont également présentes en forte quantité dans les légumes qui eux peuvent contenir moins de glucides.
- L'énergie semble fortement corrélée de manière positive à la présence de graisse et dans une moindre mesure aux sucres et glucides.
- Le nutrition-score lui est fortement corrélé de manière positive à l'énergie, aux matières grasses. Dans une moindre mesures, le sucre et le sel semblent également avoir un impact.

Afin de confirmer les observations faites sur les potentielles relations linéaires entre l'énergie et les autres données nutritionelles, nous allons effectuer une régression linéaire multiple. L'étude ne consistant pas à réaliser un modèle et pour des raisons de temps, on ne ne vérifiera pas si les hypothèses nécessaires à la réalisation d'une régression linéaire sont bien vérifiées. Les résultats sont à titre indicatifs uniquement.

```python
v='energy_100g~'
for i in nutrition.columns[1:-2]:
    v= v + '+' + i
print(v)
```

## Positionnement des marques pour une catégorie de produit et une donnée nutrionnelle définis

Ci dessous la fonction qui permettra de grouper les données nécessaires et de les tracer.
A noter que le marques seront classées par moyenne décroissante en partant de la gauche du graphique.
On effectuera aussi un test de Kruskal-Wallis pour comparer les moyennes de chaque échantillon correspondant à une marque, afin de voir si la marque a bien un impact sur la donnée nutritionelle choisie.
L'hypothèse nulle de ce test sera l'égalité des moyennes et l'hypothèse alternative la différence entre celles-ci.
Le test de Kruskal Wallis a été choisi car après plusieurs essais je me suis rendu compte que les conditions nécessaires à l'application d'une ANOVA n'étaient pas toujours remplies, notamment celle de l'égalité des variances.

```python
def select_and_plot_type_nutri(prod_cat, nutri_param, min_prod):
    
    brands = data.loc[:,['pnns_groups_2','brands',nutri_param]]
    brands['brands'] = brands['brands'].str.lower()
    selection = brands.loc[brands['pnns_groups_2'] == prod_cat]
    brands_counts = selection['brands'].value_counts() > min_prod
    brands_counts = brands_counts.loc[brands_counts==True]
    selection = selection.loc[selection['brands'].isin(list(brands_counts.index))]
    
    brand_array = selection.groupby('brands')[nutri_param].apply(lambda df: df.values)
    H, pval = kruskal(*brand_array)
    print('\n' + 'Test de Kruskal-Wallis :')
    print('Satistique H : ' + str(H))
    print('p-value : ' + str(pval))
    
    plt.figure(figsize=(25,10))
    temp = selection.groupby('brands').mean().sort_values(by=nutri_param, ascending=False)
    ax = sns.boxplot(x='brands', y=nutri_param, data=selection, order=temp.index)
    ax.set_ylabel(nutri_param, fontsize=40, labelpad =30)
    ax.set_xlabel('')
    ax.set_title('Comparison of ' + str(nutri_param[:-5]) + ' in ' + str(prod_cat), fontsize=40, pad =30)
    plt.xticks(rotation=90, fontsize=30)
    plt.yticks(fontsize=30)
    #plt.savefig( 'Comparison of ' + str(nutri_param[:-5]) + ' in ' + str(prod_cat) + '.png', format='png', bbox_inches = "tight", dpi=300)
    
    figtemp, axtemp = plt.subplots()
    temp.hist(bins=20, color='b', ax=axtemp)
    axtemp.axvline(selection[nutri_param].mean(), color='r', linewidth=2)
    #axtemp.axvline(selection[nutri_param].median(), color='g', linewidth=2)
    axtemp.set_title('Brands means distribution')
```

### Comparaison du taux de sel dans les sandwich

```python
select_and_plot_type_nutri('sandwiches','salt_100g',10)
```

On observe que la marque 'Herta' semble avoir le taux moyen de sel le plus élevé et la marque 'marks & spencer' le plus bas. Visuellement, il est difficile de voir si une marque est significativement moins salée qu'une autre. Néanmoins, la p value de l'ordre de 10^-12 obtenue lors du test de Kruskal-Wallis indique que la marque a bien un effet sur le taux de sel dans les sandwiches.

## Comparatifs entre deux marques de la moyenne d'une donnée nutritionelle

Comme vu précédemment, il peut parfois être difficile de savoir si l'une marque se positionne mieux qu'une autre en analysant juste les boxplots. Afin de voir si une marque se démarque réellemment d'une autre sur un paramètre en particulier, nous allons procéder à des tests statistiques sur les moyennes.
Ci dessous la fonction qui nous permettra d'effectuer le comparatif des moyennes entre deux marques sur une donnée nutritionelle choisie.
Celle ci effectue les opération suivante :
- Calcul du nombre de produits dans chaque échantillon
- Calcul de la moyenne et de l'écart type de chaque échantillon
- Affichage de la distribution de chaque échantillon ainsi que sa moyenne
- Vérification de la normalité de la distribution (condition pour test de student)
- Vérification de l'égalité des variances de Levene (condition pour test de student)
- Test sur les moyenne avec un test de Student
- Test sur les moyennes avec un test de Mann Whitney

A noter qu'ici, après plusieurs essais préalables j'ai pu me rendre compte que les conditions pour le test de student ne sont pas toujours remplies. Il est donc effectué à titre indicatif (car possiblement robuste à la violation de certaines de ses hypothèses), néanmoins seul le résultat du test non paramétrique de Mann Whitney est réellement pris en compte.
Pour le test de Student et celui de Mann Whitney, l'hypothèse H0 est l'égalité des moyennes et l'hypothèse alternative H1 la supériorité de la moyenne de la première marque par rapport à la seconde.

```python
def mean_compare(prod_cat, nutri_param, brand_1, brand_2):
    """
        Compare la moyenne de deux échantillons de deux marques issu d'une même catégorie de produit
        sur un paramètre nutritionnel particulier.
        prod_cat : catégorie de produit à sélectionner (string)
        nutri_param : paramètre nutrionnel sur lequel effectuer la comparaison (string)
        brand_1 : nom de la première marque à choisir (string)
        brand_2 : nom de la seconde marque à choisir (string)
    """
    
    print('Comparaison des marques ' + str(brand_1) + ' et ' + str(brand_2) 
          + ' pour la variable ' + str(nutri_param))
    
    brands = data.loc[:,['pnns_groups_2','brands',nutri_param]]
    brands['brands'] = brands['brands'].str.lower()
    selection = brands.loc[brands['pnns_groups_2'] == prod_cat]
    selection_b1 = selection.loc[selection['brands'] == brand_1]
    selection_b2 = selection.loc[selection['brands'] == brand_2]
    print('\n' + 'Nombre de produits dans chaque échantillon :')
    print(brand_1 + ' : ' + str(len(selection_b1)))
    print(brand_2 + ' : ' + str(len(selection_b2)))

    print('\n' + 'Moyenne et écart type pour chaque échantillon :')
    m_b1 = np.mean(selection_b1[nutri_param])
    m_b2 = np.mean(selection_b2[nutri_param])
    dev_b1 = np.std(selection_b1[nutri_param], ddof=1)
    dev_b2 = np.std(selection_b2[nutri_param], ddof=1)
    print('Pour la marque ' + str(brand_1))
    print('moyenne : ' + str(m_b1))
    print('écart type : ' + str(dev_b1))
    print('Pour la marque ' + str(brand_2))
    print('moyenne : ' + str(m_b2))
    print('écart type : ' + str(dev_b2))
    
    fig, ax = plt.subplots(2,1, sharex=True, figsize=(10,10))
    selection_b1.hist(figsize=(5,5), bins=20, color='b', ax=ax[0])
    selection_b2.hist(figsize=(5,5), bins=20, color='b', ax=ax[1])
    ax[0].axvline(m_b1, color='r', linewidth=2)
    ax[1].axvline(m_b2, color='r', linewidth=2)
    ax[0].set_title(brand_1, fontsize=20, pad=10)
    ax[1].set_title(brand_2, fontsize=20, pad=10)
    ax[1].set_xlabel(nutri_param, labelpad=20, fontsize=15)
    ax[0].tick_params(labelsize=15)
    ax[1].tick_params(labelsize=15)
    #plt.savefig( 'Comparison ' + str(brand_1) + ' & ' + str(brand_2)+ '.jpeg', format='jpeg', bbox_inches = "tight")
    
    print('\n' + 'Test de normalité de Kolmogorov-Smirnov :')
    
    ks_b1 = ks_2samp(selection_b1[nutri_param],list(np.random.normal(m_b1, dev_b1, 1000)))
    ks_b2 = ks_2samp(selection_b2[nutri_param],list(np.random.normal(m_b2, dev_b2, 1000)))
    print('Pour la marque ' + str(brand_1))
    print('Satistique ks : ' + str(ks_b1[0]))
    print('p-value : ' + str(ks_b1[1]))
    print('Pour la marque ' + str(brand_2))
    print('Satistique ks : ' + str(ks_b2[0]))
    print('p-value : ' + str(ks_b2[1]))
    
        
    print('\n' + 'Test de Levene sur les variances :')
    l_test = levene(selection_b1[nutri_param], selection_b2[nutri_param])
    print('Satistique W : ' + str(l_test[0]))
    print('p-value : ' + str(l_test[1]))
    
    print('\n' + 'Test de Student sur les moyennes :')
    
    if ((len(selection_b1) <= 20) or (len(selection_b1) <= 20)):
        print('Attention au moins un des échantillons contient moins de 30 individus')
        
    if (selection_b1[nutri_param].mean() >= selection_b2[nutri_param].mean()):
        H1 = 'larger'
    else:
        H1 = 'smaller'
        
    t_test = sm.stats.ttest_ind(selection_b1[nutri_param], selection_b2[nutri_param], alternative=H1)
    print('Satistique t : ' + str(t_test[0]))
    print('p-value : ' + str(t_test[1]))
    
    print('\n' + 'Test de Mann Whitney sur les moyennes:')
    
    if ((len(selection_b1) <= 20) or (len(selection_b1) <= 20)):
        print('Attention au moins un des échantillons contient moins de 20 individus'
             )
    if (selection_b1[nutri_param].mean() >= selection_b2[nutri_param].mean()):
        H1 = 'greater'
    else:
        H1 = 'less'
        
    u_test = mannwhitneyu(selection_b1[nutri_param], selection_b2[nutri_param], use_continuity=True, 
                 alternative=H1)
    print('Satistique U : ' + str(u_test[0]))
    print('p-value : ' + str(u_test[1]))
```

### Comparaison du taux de matières grasses dans les appéritifs Pringles et U

```python
mean_compare('appetizers', 'fat_100g', 'pringles', 'u')
```

Visuellement il est flagrant que la marque Pringles a tendance à être plus grasse que la marque U. Cela se vérifie avec la p-value du test de Mann Whitney qui est de l'ordre de 10-13.
